## Task 2.....

```python
### add 5 grace marks to each student's marks
grace_marks_series = marks_series + 5
grace_marks_series
```

```python
### subtract 2 marks from each student's marks
grace_marks_series = grace_marks_series - 2
grace_marks_series
```

```python
### multiply each student's marks by 1.05
grace_marks_series = grace_marks_series * 1.05
grace_marks_series
```

```python
### devide each student's marks by 2
grace_marks_series = grace_marks_series / 2
grace_marks_series
```

## Task 3.....

```python
marks_series
```

```python
## maximum marks
marks_series.max()
```

```python
## minimum marks
marks_series.min()
```

```python
## Sum of marks
marks_series.sum()
```

```python
## mean of marks
marks_series.mean()
```

```python
## check student passed or failed (pass mark is 70)
pass_fail_series = marks_series.apply(lambda x: 'Pass' if x >= 70 else 'Fail')
pass_fail_series
```

```python
### count passed students
val_count = pass_fail_series.value_counts()
val_count['Pass']
```

## Task 4....

```python
students = {
    'name': ['Amit', 'Neha', 'Rahul', 'Sneha', 'Pooja'],
    'marks': [78, 85, 90, 66, 72],
    'subject': ['Math', 'Math', 'Science', 'Science', 'Math']
}
students_df = pd.DataFrame(students)
students_df
```

```python
## Students by marks in descending order
students_df.sort_values(by='marks', ascending=False, inplace=True)
students_df
```

```python
### reset index after sorting
students_df.reset_index(drop=True, inplace=True)
students_df
```

## Task 6.....

```python
students_df[students_df['subject'] == 'Math']
```

## Task 7.....

```python
## find average marks per subject using groupby
students_df.groupby('subject')['marks'].mean()
```
